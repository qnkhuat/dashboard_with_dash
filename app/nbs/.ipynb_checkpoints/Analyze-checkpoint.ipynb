```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.plotly as py
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
plt.rcParams['figure.figsize'] = (10,8)
```

### Load data

```python

df_u = queries.df_user()

# find id of user that are not suppose to has
exclude_id = []
for name in ['gv','test','alec','dyon tran','khoa']:
    exclude_id.extend(df_u[df_u.fullname.str.lower().str.contains(name)].id)
df_cl = queries.df_classinfo()
# find classes taught by these users
exclude_classid = df_cl[df_cl.teacherid.isin(exclude_id)].roomid
df_cl = df_cl[~df_cl.roomid.isin(exclude_classid)].reset_index()
df_e = pd.read_csv('../src/data/emotion_23042019.csv')
df_e = df_e[~df_e.id_class.isin(exclude_classid)].reset_index()

df_e_cl = df_e.merge(df_cl,left_on='id_class',right_on='roomid')
assert df_e_cl.shape[0]==df_e.shape[0] , "Shape of Dataframe changed"


df_cl_e = df_cl.merge(df_u,left_on='studentid',right_on='id',how='left',suffixes=('','_kid')).copy()
df_cl_e = df_cl_e.merge(df_u,left_on='teacherid',right_on='id',how='left',suffixes=('','_teacher')).copy()
df_cl_e['emotion_score_teacher'] = df_cl_e.roomid.map(df_e[df_e.user_role=='teacher'].groupby('id_class').emotion_score.mean().to_dict())
df_cl_e['emotion_score_kid'] = df_cl_e.roomid.map(df_e[df_e.user_role=='student'].groupby('id_class').emotion_score.mean().to_dict())
df_cl_e['emotion_score'] = df_cl_e.roomid.map(df_e.groupby('id_class').emotion_score.mean().to_dict())
assert df_cl_e.shape[0]==df_cl.shape[0], "Shape of Dataframe changed"
```

### Helpers

```python
def uinfo(id): return df_u[df_u.id==id]
```

### Emotion score vs starttime

```python
df_cl_e.starttime.dt.hour.value_counts(sort=False).plot(kind='bar')
plt.title('Classes number through time')
```

```python

data = []
for col in ['emotion_score_kid','emotion_score_teacher','emotion_score']:
    df = df_cl_e.groupby(df_cl_e.starttime.dt.hour)[col].mean()
    trace = go.Bar(
        x=df.index,
        y=df.values,
        text = df_cl_e.starttime.dt.hour.value_counts(sort=False),
        name=col)
    data.append(trace)
f = go.Figure(data=data,layout=go.Layout(width=700,height=500))
pyo.iplot(f)
```

### Emotion rank

```python
d1 = []
for col in ['emotion_score_kid','emotion_score_teacher','emotion_score']:
    d1.append(go.Scatter(x=df_cl_e.starttime,
         y=df_cl_e[col],
         mode='markers',
         marker = {'opacity':0.5},
         text=df_cl_e.roomid,
         name=col))
l1 = go.Layout(
    hovermode='closest',
    width=700,
    height=500,)
pyo.iplot(go.Figure(d1,l1))
```

### Male vs Female teachers / kids

```python
d2 = []
for col in ['male','female','class']:
    sex = [col] if col!='class' else ['male','female']
    trace = go.Bar(
        x=['emotion_score_kid','emotion_score_teacher','emotion_score'],
        y=df_cl_e[df_cl_e.kid_sex.isin(sex)][['emotion_score_kid','emotion_score_teacher','emotion_score']].mean(),
        name=col
    )
    d2.append(trace)
l2 = go.Layout(
    hovermode='closest',
    width=700,
    height=500,)
pyo.iplot(go.Figure(d2,l2))
```

```python
rand_safe = (np.random.rand(20,2)-.5)/10
rand_danger = (np.random.rand(20,2)-.5)/10
```

```python
miami = [25.761681,-80.191788]
miami_shift = [25.761681,-80.291788]
```

```python
safer = rand_safe + miami_shift 
indanger = rand_danger + miami_shift 

datasets = np.concatenate([safer,indanger])

from sklearn.cluster import KMeans

model = KMeans(n_clusters=4)
model.fit(datasets)

centroid = model.cluster_centers_

data = [
    go.Scattermapbox(
        name='Safer',
        lat=safer[:,0],
        lon=safer[:,1],
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=9
        ),
        text=["The coffee bar","Bistro Bohem","Black Cat",
             "Snap","Columbia Heights Coffee","Azi's Cafe",
             "Blind Dog Cafe","Le Caprice","Filter",
             "Peregrine","Tryst","The Coupe",
             "Big Bear Cafe"],
    ),go.Scattermapbox(
        name='Indanger',
        lat=indanger[:,0],
        lon=indanger[:,1],
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=9
        ),
        text=["The coffee bar","Bistro Bohem","Black Cat",
             "Snap","Columbia Heights Coffee","Azi's Cafe",
             "Blind Dog Cafe","Le Caprice","Filter",
             "Peregrine","Tryst","The Coupe",
             "Big Bear Cafe"],
    ),
    go.Scattermapbox(
        name='Safe place',
        lat=centroid[:,0],
        lon=centroid[:,1],
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=15,
            color = 'rgb(255, 0, 0)'
        ),
        text=["Safe place {}".format(index) for index in range(len(centroid))]
    ),
    go.Scattermapbox(
        name='Hurricane',
        showlegend=False,
        lat=[25.71410567],
        lon=[-80.18339028],
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=15,
            color = 'rgb(255, 0, 0)',
            symbol='circle-stroked'
        
        ),
        text=["Safe place {}".format(index) for index in range(len(centroid))]
    ),
    
]

layout = go.Layout(
    autosize=True,
#     showlegend=False,
    paper_bgcolor='rgb(56, 56, 56)',
    hovermode='closest',
    legend = {
        'bgcolor':'rgb(255,255,255)'
    },
    mapbox=go.layout.Mapbox(
        accesstoken='pk.eyJ1IjoibGFtaG9hbmd0dW5nIiwiYSI6ImNqdjF3OTZxbDB4YzMzeWtjanc2ZmNwemwifQ.m0zuohyUWthY8aAIc82_Yg',
        bearing=0,
        center=go.layout.mapbox.Center(
            lat=miami[0],
            lon= miami[1]
        ),
        pitch=10,
        zoom=10
    ),
#     images= [dict(
#         source= "https://raw.githubusercontent.com/qnkhuat/scripts/master/hurricane.png",
#         x= 0.6,
#         y= 0.75,
#         sizex= .09,
#         sizey= .05,
#         xref='x',
#         yref='y',
#         layer="below",
#         sizing="stretch"
#       )
#     ]
)

fig = go.Figure(data=data, layout=layout)
py.plot(fig, filename='Multiple Mapbox')
```
